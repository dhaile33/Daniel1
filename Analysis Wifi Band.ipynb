```python
import h5py
import numpy as np
import pandas as pd
```

```python
name = 'IITSO_20190131000852_(30-6000_MHz).h5'
f = h5py.File(name, 'r')
```

```python
number = '10'
dset = f['sessions']['0']['spectrum'][number]
```

```python
start = int(dset.attrs.get('start_freq'))/1000000
stop = int(dset.attrs.get('stop_freq'))/1000000
amount = int(dset.attrs.get('num_points'))
noisefloor=int(dset.attrs.get('ref_level'))
noisefloor
```

```python
timestamps = dset['start_time']
```

```python
powers = dset['powers']
```

```python
frequencies = np.linspace(start, stop, amount)
```

```python
df = pd.DataFrame(powers[0:], timestamps, frequencies)
df
```

```python
df.to_csv('Daniel10Jan2018.csv')
```

```python
df.to_hdf('Daniel10Jan2018.h5', key='df', mode = 'w')
```

```python
pd.read_hdf('Daniel10Jan2018.h5', 'df')
```

```python
number = '20'
dset = f['sessions']['0']['spectrum'][number]

start = int(dset.attrs.get('start_freq'))
stop = int(dset.attrs.get('stop_freq'))
amount = int(dset.attrs.get('num_points'))

timestamps = dset['start_time']

powers = dset['powers']

frequencies = np.linspace(start, stop, amount)

df = pd.DataFrame(powers[0:], timestamps, frequencies)
```

```python
df.to_csv('Daniel20Jan2018.csv')
```

```python
df.to_hdf('Daniel20Jan2018.h5', key='df', mode = 'w')
```

```python
pd.read_hdf('Daniel20Jan2018.h5', 'df')
```
